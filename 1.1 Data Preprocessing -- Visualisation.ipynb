# Data Preprocessing: Visualisation

The first step to developing a machine learning model is to examine your data. Data examination and visualisation is important to highlight how features correlate, what type of variable each feature is (string, int, etc), and how many missing values are present. Depending on the dataset size and number of features, there are a few common strategies to help get an understanding of the set.

## Initial Printout

In a smaller dataset, a good approach is simply to print the set out:

```python
import pandas as pd

df = pd.read_csv("titanic/train.csv")
df.to_string
```

In very large datasets, it can be impractical to load and display all the rows. You can instead call df.head() to print the first five rows, and df.info() to show the column type and number of non-null values per column.

```python
df.head()
```

```python
df.info()
```

From here, we can note a few things:
1) Some columns such as Cabin contain many missing values. If we were to indiscriminately filter out all rows with missing values, we would only be left with 204 rows, eliminating over 75% of the data.
2) Some columns contain unique values for every row. For example, every name in the dataset is unique. If we just plugged the Name column in as a feature to an ML model, the model wouldn't be able to draw any meaningful conclusions from that feature. This doesn't mean, however, that the column is useless. Information such as a person's title may correlate to whether or not they survived (we'll explore this later).
3) Some values will need to be converted to be useful. For example, the Sex column can easily be converted to be of type float, by assigning 'male' = 0.0, and 'female' = 1.0. This is useful information for when we begin cleaning the data.
